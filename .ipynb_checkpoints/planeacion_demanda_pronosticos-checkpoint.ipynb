## 1.2. Pre-procesamiento de datos

### 1.2.1 Convertir Periodo a Semana

```python
# Asignar formato datetime a columna Fecha
df['FECHA'] = pd.to_datetime(df['FECHA'], format='%d-%m-%y')

# Establecer FECHA como datetime index
df.set_index('FECHA', inplace=True)

# Eliminar columna COD_ALMACEN
df = df.drop(columns = 'COD_ALMACEN')

# Filtrar por fechas posteriores a 2021-01-01
df = df[df.index >= '2021-01-01'].copy()

# Usar función GROUP BY para agrupar demanda por semana, comenzando los lunes y terminando los domingos
df_sem = df.groupby(['COD_SKU', 'DESC_SKU']).resample('W-SUN').sum(numeric_only=True)

# Resetear Index para aplanar la tabla
df_sem.reset_index(inplace=True)

# Colocar semanas como columnas con una tabla dinamica
df_sem_td = df_sem.pivot(index=['COD_SKU', 'DESC_SKU'], columns='FECHA', values='DEMANDA').fillna(0)

df_sem_td
```

### 1.2.2 Construir ABC

```python
# Carga la informacion de Costos Unitarios y Utilidad Unitaria
ruta_costos = r'dataset\sku_costos_utilidad.xlsx'
df_skus_costos = pd.read_excel(ruta_costos)
```

```python
# Selecciona las nombres de columna de las ultimas 52 semanas
demanda_ultimo_año = df_sem_td.columns[-52:].tolist()

# Suma las ultimas 52 semanas (un año) de demanda y crea una columna llamada 'TOTAL DEMANDA'
df_sem_td['TOTAL_DEMANDA'] = df_sem_td[demanda_ultimo_año].sum(axis=1)

# Selecciona las columnas a traer 
columnas_importar = ['COD_SKU', 'DESC_SKU', 'COSTO_UNITARIO','UTILIDAD_UNITARIA', 'PRECIO',	'INVENTARIO_ACTUAL']

# Trae las columnas correspondientes a cada codigo. Es equivalente a un BuscarV/BuscarX
dem_sem = df_sem_td.merge(df_skus_costos[columnas_importar], on=['COD_SKU','DESC_SKU'], how ='left')

# Calcular total utilidad
dem_sem['TOTAL_UTILIDAD'] = dem_sem['TOTAL_DEMANDA'] * dem_sem['UTILIDAD_UNITARIA']

# Ordenar Descendentemente por 'Total_Utilidad'
dem_sem = dem_sem.sort_values(by='TOTAL_DEMANDA', ascending=False)

# Calcula y adiciona la participación individual de cada sku 
dem_sem['PART_IND'] = dem_sem['TOTAL_UTILIDAD']/dem_sem['TOTAL_UTILIDAD'].sum()

# Calcula la participacion acumulada 
dem_sem['PART_ACUM'] = dem_sem['PART_IND'].cumsum()
```

```python
def ABC_segmentacion(part): #función para aplicar a cada valor de participacion acumulada
   
    if part > 0 and part < 0.8:
        return 'A'
    elif part >= 0.80 and part < 0.95:
        return 'B'
    elif part >= 0.95:
        return 'C'
    
# Aplica la funcion de ABC_segmentacion a cada porcentaje acumulado
dem_sem['CLASIF_ABC'] = dem_sem['PART_ACUM'].apply(ABC_segmentacion)
```

```python
# Resetear Indice y eliminar columna gnerada (index) por defecto
dem_sem = dem_sem.reset_index().drop(columns = 'index')
dem_sem
```

## 1.3. Funciones para calcular PMS

```python
# Seleccionar las columnas que comienzan por '202' (las de demanda)
columnas_dem = dem_sem.filter(like='202')

# Llevar valores de demanda a una lista
val_dem_sem = columnas_dem.values.tolist()
```

### 1.3.1 Calculo de PMS por SKU de forma individual

```python
def promedio_movil(demanda, extra_periods, n):
    
    # Determina el numero de datos de la demanda
    largo_demanda = len(demanda) 
    # Adiciona el numero de periodos que queremos pronosticar
    demanda = np.append(demanda, [np.nan] * extra_periods)
    # Crea un arreglo para posteriormente escribir los pronosticos de largo demanda + los perdiodos extras
    forecast = np.full(largo_demanda + extra_periods, np.nan)  
    
    # Itera entre n y el largo de la serie de demanda
    for t in range(n, largo_demanda):
        # Promedia los datos de la demanda correspondientes segun n y los datos disponibles
        forecast[t] = np.mean(demanda[t-n:t]) 
    for u in range(1,extra_periods + 1):
        # Pronostica el periodo siguiente
        forecast[t+u] = np.mean(demanda[t-n+1:t+1]) 
    
    # Crea el data frame con los pronósticos
    df = pd.DataFrame.from_dict({'DEMANDA': demanda, 'FORECAST': forecast, 'ERROR': demanda-forecast,}) 
    
    # Calcula 'Error Porcentual' usando función lambda 
    df['ERROR_PORCENTUAL'] = df.apply(
                    lambda fila: 2 if (fila['DEMANDA'] <= 0 and pd.notna(fila['FORECAST'])) else abs(fila['DEMANDA'] 
                        - fila['FORECAST']) / fila['DEMANDA'], axis=1)
  
    # Regresa df como resultado   
    return df
```

### 1.3.2 Funcion para calcular KPIs

```python
def kpi(df):
    #Grafica
    # Nombra el indice como 'Periodo'
    df.index.name = 'Periodo'
    # Aplicar estilo 'ggplot'
    plt.style.use('ggplot')
    # Grafica la demada y el pronostico
    df[['DEMANDA','FORECAST']].plot(figsize=(8,3),title='Demanda vs Pronostico', style=['-','--']) 
    # Muestra la grafica
    plt.tight_layout()
    # Suma la demanda de las filas que tienen dato de error (donde 'Error' no es NaN)
    suma_demanda = df.loc[df['ERROR'].notnull(), 'DEMANDA'].sum() 
    # Calcula el wmape sumando el valor absoluto de los datos de error y
    # Dividiendo por la suma de la demanda correspondiente. Coloca 2 (200%) si la suma de la demanda es 0
    error_abs = df['ERROR'].abs().sum()
    mae_porc = 2 if suma_demanda <= 0 else error_abs / suma_demanda  
    
    # Muestra el wmape
    print('MAE%: {:.2%}'.format(mae_porc)) 
    # Calcula el sesgo%
    sesgo = (lambda x: 2 if x <= 0 else df['ERROR'].sum() / x)(suma_demanda)
    # Muestra el sesgo%
    print('Sesgo: {:.2%}'.format(sesgo)) 
    
    # Calcula el score
    score = mae_porc + abs(sesgo)
    # Muestra el score
    print('Score: {:.2%}'.format(score))
    
```

### 1.3.3 Aplicación PMS por Sku

### 1.3.5 Funcion para optimizar el numero de periodos n por cada Sku

```python
def optimizacion_promedio_movil(demanda, extra_periods):
    
    # Crea una lista vacia para acumular los datos de n
    params = []
    # Crea una lista vacia para acumular los datos de error
    KPIs = [] 
    # Crea una lista vacia para acumular las tablas de pronostico de cada posible n
    dfs = [] 
    
    # Itera entre n's diferentes desde 1 hasta 50 (para que queden al menos 3 datos de error)
    for n in range(1,50):
        # Crea el pronostico conla funcion promedio_movil
        df_prom_mov =  promedio_movil(demanda, extra_periods = extra_periods, n = n)
        # Acumula el numero de periodos n
        params.append(f'Promedio movil, n={n}') 
        # Acumula la tabla completa
        dfs.append(df_prom_mov)
        # Calcula la suma de la demanda para cade n 
        suma_demanda = df_prom_mov.loc[df_prom_mov['ERROR'].notnull(), 'DEMANDA'].sum()
        # Calcula el wmape para cada n, devuelve 2 (200%) si la suma de la demanda es 0
        mae_porc = 2 if suma_demanda <= 0 else df_prom_mov['ERROR'].abs().sum() / suma_demanda
        # Acumula el error en la lista KPI
        KPIs.append(mae_porc)
        
    # Selecciona el minimo de cada n    
    minimo = np.argmin(KPIs)
    # Muestra el n minimo y el error correspondiente
    print(f'Mejor {params[minimo]} con WMAPE de', "{:.2%}".format(KPIs[minimo]))
    # Da formato % 
    formatted_KPIs = ["{:.2%}".format(error) for error in KPIs]
    print('Resultado errores:',formatted_KPIs)
   
    return dfs[minimo]
```

### 1.3.6 Aplicación de funcion de optimización por Sku

```python
# Ejecuta funcion de optimizacion
n_opt = optimizacion_promedio_movil(demanda = val_dem_sem[2] , extra_periods=1)
display(n_opt)
kpi(n_opt)
```

### 1.3.7 Función para aplicar optimización a todos los Sku

```python
def generacion_mejor_promedio_movil(series_tiempo, extra_periods, n_min, n_max):
    
    # Crea una lista vacia para acumular el pronostico del periodo siguiente 
    forecast_siguiente = [] 
    # Crea una lista vacia para acumular el wmape correspondiente de cada referencia
    mejor_error = []
    # Crea una lista vacia para acumular el n con menor error
    mejor_n = []
    # Crea una lista vacia para acumular el n con menor error
    rmse_mejor_n = []

    # Para calculo del error total
    total_error_abs = []
    total_demanda = []
    
    # Itera por cada una de las series de tiempo
    for serie in series_tiempo: 
        params = []
        KPIs = []
        dfs = []
        RMSE = []
        error_abs = []
        suma_demanda = []
        
        for n in range(n_min,n_max):        
            # Aplica funcion de promedio por cada n
            df_prom_mov =  promedio_movil(serie, extra_periods = extra_periods, n = n)
            # Acumula los parametros
            params.append(f'Promedio movil, n: {n}')
            # Acumula las tablas
            dfs.append(df_prom_mov)
            # Suma la demanda que aplica para calcular el error
            sum_dem = df_prom_mov.loc[df_prom_mov['ERROR'].notnull(), 'DEMANDA'].sum()
            sum_len = df_prom_mov.loc[df_prom_mov['ERROR'].notnull(), 'DEMANDA'].count()
            sum_error_abs = df_prom_mov['ERROR'].abs().sum()
            # Calcula el wmape, devuelve 200% si demanda igual o menor que 0
            mae_porc = 2 if sum_dem <= 0 else sum_error_abs / sum_dem
            rmse = np.sqrt((df_prom_mov['ERROR']**2).sum() / sum_len)
            # Acumula KPI
            KPIs.append(mae_porc)
            RMSE.append(rmse)
            error_abs.append(sum_error_abs)
            suma_demanda.append(sum_dem)
            
        # Seleeciona el minimo de cada n
        minimo = np.argmin(KPIs)
        # Selecciona el wmape minimo
        mae_porc_minimo = np.min(KPIs)
        # Selecciona la suma del error absoluto correspondiente al mejor n 
        sum_error_abs_min = error_abs[minimo]
        # Selecciona la suma de la demanda correspondiente al mejor n 
        sum_dem_min = suma_demanda[minimo]
        # Selecciona el rmse correspondiente al mejor n (no necesariamente es el menor rmse)
        rmse_n = RMSE[minimo]
        # Selecciona el pronostico de la tabla con mejor wmape
        forecast =  dfs[minimo]['FORECAST'].iloc[-1]
        # Acumula el pronostico
        forecast_siguiente.append(forecast)
        # Acumula el wmape
        mejor_error.append(mae_porc_minimo)
        # Acumula el mejor n con base en el indice +1
        mejor_n.append(minimo+1)
        # Acumula el rmse correspondiente al mejor n
        rmse_mejor_n.append(rmse_n)
        #Calcula error global
        total_error_abs.append(sum_error_abs_min)
        total_demanda.append(sum_dem_min)

    # Calcula el mae% de todos los sku
    error_global = np.sum(total_error_abs) / np.sum(total_demanda) if np.sum(total_demanda) != 0 else float('inf')
    
    return forecast_siguiente, mejor_error, mejor_n, rmse_mejor_n, error_global
```

### 1.3.8 Aplicación de la optimización de n para todos los Sku

```python
#Calcula el mejor n para cada SKU - puede tomar 1 minuto
forecast_siguiente, mejor_error, mejor_n, rmse_mejor_n, error_global  = generacion_mejor_promedio_movil(
                                                                            series_tiempo=val_dem_sem, 
                                                                            extra_periods=1, 
                                                                            n_min=1, 
                                                                            n_max=53) 
print('MAE% Global:', "{:.2%}".format(error_global))
```

```python
# Funcion para clasificar XYZ
def XYZ_segmentation(mae_porc):
    if mae_porc <= 0:
        return 'Z'
    elif mae_porc > 0 and mae_porc <= 0.5:
        return 'X'
    elif mae_porc > 0.5 and mae_porc <= 1:
        return 'Y'
    elif mae_porc > 1:
        return 'Z'

# Asignar Clasificacion XYZ
dem_sem['CLASIF_XYZ'] = dem_sem['MEJOR_MAE%'].apply(XYZ_segmentation)

dem_sem
```

# <b><font color="navy">2. Suavización Exponencial SE</span></font></b>

### 2.1. Funcion para calcular SE Simple por cada Sku

```python
def suavizacion_exp_simple(demanda, periodos_extra, alfa):

    # periodos historicos
    largo_demanda = len(demanda)

    # Acumular np.nan en el arreglo de demanda para cubrir periodos  futuros
    d = np.append(demanda, [np.nan]*periodos_extra)

    # Arreglo para el pronostico
    f = np.full(largo_demanda +  periodos_extra, np.nan)
    # Inicializar modelo
    f[1] = d[0]

    #Crear todos los pronosticos t+1 hasta el final de los periodos historicos
    for t in range(2, largo_demanda+1):
        f[t] = alfa*d[t-1]+(1-alfa)*f[t-1]

    # Pronostico para los pediodos extra
    for t in range(largo_demanda+1, largo_demanda+periodos_extra):
        # Actualizar el pronostico con el pronostico anterior
        f[t] = f[t-1]

    df = pd.DataFrame.from_dict({'DEMANDA':d,'FORECAST':f,'ERROR':d-f})
    
    return df
    
```

```python
# Seleccionar SKU indicando el index [0, 1, 2..]
demanda = val_dem_sem[5]

# Indicar periodos extra a pronosticar
periodos_extra = 2

# Indicar valor de Alfa
alfa = 0.3

# Aplicar Funcion SE
suav_exp = suavizacion_exp_simple(demanda, periodos_extra, alfa)

# Mostrar resultados
kpi(suav_exp)
suav_exp
```

### 2.2 Funcion para optimizar alfa por cada Sku

```python
def suavizacion_exp_optimo(demanda, periodos_extra, rango_alfa):
    parametros = []
    KPIs = []
    dfs= []

    for alfa in rango_alfa:
    
        df = suavizacion_exp_simple(demanda, periodos_extra = periodos_extra, alfa = alfa)
    
        parametros.append(f'Suavizacion Exponencial Simple, alfa:{alfa}')
        dfs.append(df)
        
        suma_demanda = df.loc[df['ERROR'].notnull(), 'DEMANDA'].sum()
        # Calcula el wmape para cada n, devuelve 2 (200%) si la suma de la demanda es 0
        mae_porc = 2 if suma_demanda <= 0 else df['ERROR'].abs().sum() / suma_demanda
        # Acumula el error en la lista KPI
        KPIs.append(mae_porc)

     # Selecciona el minimo de cada n    
    minimo = np.argmin(KPIs)
    # Muestra el n minimo y el error correspondiente
    print(f'Mejor {parametros[minimo]} con MAE% de', "{:.2%}".format(KPIs[minimo]))
    # Da formato % 
    formatted_KPIs = ["{:.2%}".format(error) for error in KPIs]
    print('Resultado errores:',formatted_KPIs)
   
    return dfs[minimo]
```

```python
# Seleccionar SKU indicando el index [0, 1, 2..]
demanda = val_dem_sem[0]

# Indicar periodos extra a pronosticar
periodos_extra = 2

# Seleccionar rango para optimizar alfa
rango_alfa = np.arange(0.01, 0.61, 0.01)

# Ejecutar funcion optimizacion
suavizacion_exp_optimo(demanda=demanda, periodos_extra=periodos_extra, rango_alfa=rango_alfa)
```

### 2.3 Función para aplicar mejor alfa a todos los Sku

```python
def generacion_mejor_suavizacion_exp(series_tiempo, periodos_extra, rango_alfa):
    # Crea una lista vacia para acumular el pronostico del periodo siguiente 
    forecast_siguiente = [] 
    # Crea una lista vacia para acumular el wmape correspondiente de cada referencia
    mejor_error = []
    # Crea una lista vacia para acumular el alfa con menor error
    mejor_alfa = []
    # Crea una lista vacia para acumular el alfa con menor error
    rmse_mejor_alfa = []

    # Para calculo del error total
    total_error_abs = []
    total_demanda = []
    
    # Itera por cada una de las series de tiempo
    for serie in series_tiempo: 
        parametros = []
        KPIs = []
        dfs = []
        RMSE = []
        error_abs = []
        suma_demanda = []
        
        for alfa in rango_alfa:
                  
            # Aplica funcion de promedio por cada alfa
            df =  suavizacion_exp_simple(demanda=serie, periodos_extra = periodos_extra, alfa = alfa)
            # Acumula los parametros
            parametros.append(alfa)
            # Acumula las tablas
            dfs.append(df)
            # Suma la demanda que aplica para calcular el error
            sum_dem = df.loc[df['ERROR'].notnull(), 'DEMANDA'].sum()
            sum_len = df.loc[df['ERROR'].notnull(), 'DEMANDA'].count()
            sum_error_abs = df['ERROR'].abs().sum()
            # Calcula el wmape, devuelve 200% si demanda igual o menor que 0
            mae_porc = 2 if sum_dem <= 0 else sum_error_abs / sum_dem
            rmse = np.sqrt((df['ERROR']**2).sum() / sum_len)
            # Acumula KPI
            KPIs.append(mae_porc)
            RMSE.append(rmse)
            error_abs.append(sum_error_abs)
            suma_demanda.append(sum_dem)
            
        # Seleeciona el minimo de cada alfa
        minimo = np.argmin(KPIs)
        # Selecciona el alfa correspondiente al menor error
        mejor_param_alfa = parametros[minimo]
        # Selecciona el wmape minimo
        mae_porc_minimo = np.min(KPIs)
        # Selecciona la suma del error absoluto correspondiente al mejor n 
        sum_error_abs_min = error_abs[minimo]
        # Selecciona la suma de la demanda correspondiente al mejor n 
        sum_dem_min = suma_demanda[minimo]
        # Selecciona el rmse correspondiente al mejor n (no necesariamente es el menor rmse)
        rmse_alfa = RMSE[minimo]
        # Selecciona el pronostico de la tabla con mejor wmape
        forecast =  dfs[minimo]['FORECAST'].iloc[-1]
        # Acumula el pronostico
        forecast_siguiente.append(forecast)
        # Acumula el wmape
        mejor_error.append(mae_porc_minimo)
        # Acumula el mejor n con base en el indice +1
        mejor_alfa.append(mejor_param_alfa)
        # Acumula el rmse correspondiente al mejor n
        rmse_mejor_alfa.append(rmse_alfa)
        #Calcula error global
        total_error_abs.append(sum_error_abs_min)
        total_demanda.append(sum_dem_min)
        
    error_global = np.sum(total_error_abs) / np.sum(total_demanda) if np.sum(total_demanda) != 0 else float('inf')
    
    return forecast_siguiente, mejor_error, mejor_alfa, rmse_mejor_alfa, error_global, dfs
```

```python
# Seleccionar rango para optimizar alfa
rango_alfa = np.arange(0.01, 0.61, 0.01)

#Calcula el mejor alfa para cada SKU - puede tomar 1 minuto
forecast_siguiente_se, mejor_error_se, mejor_alfa, rmse_mejor_alfa, error_global, dfs  = generacion_mejor_suavizacion_exp(
                                                                                                series_tiempo=val_dem_sem, 
                                                                                                periodos_extra=1, 
                                                                                                rango_alfa=rango_alfa)

# Mostrar MAE%
print('MAE% Global:', "{:.2%}".format(error_global))
```

## 2.4 Suavización Exponencial con `statsmodels`

```python
from statsmodels.tsa.holtwinters import SimpleExpSmoothing, ExponentialSmoothing
```

#### 2.4.1 Suavizacion Exponencial Simple

```python
# Seleccionar Sku
demanda = val_dem_sem[5]

# Aplicar suavizacion exponencial simple
model = SimpleExpSmoothing(demanda).fit(smoothing_level=None, optimized=True)

# Entrenar el modelo
ses = model.fittedvalues

# Generar proximo pronostico
forecast_next = model.forecast(steps=1).item()

# Mostrar parametros alfa, beta, gamma
print('alfa optimo:', model.params['smoothing_level'].round(2), 'Pronostico siguiente:',np.round(forecast_next,0))
```

```python
df_ses = pd.DataFrame({'DEMANDA':demanda,'FORECAST':ses, 'ERROR':demanda-ses})
df_ses
```

```python
kpi(df_ses)
```

#### 2.4.2 Suavizacion Exponencial Triple

```python
# Seleccionar Sku
demanda = val_dem_sem[5]

# Aplicar suavizacion exponencial simple
model = ExponentialSmoothing(demanda, seasonal='add', trend='add', seasonal_periods=52).fit()

# Entrenar modelo
set = model.fittedvalues

# Mostrar parametros
print("Alfa:", model.model.params['smoothing_level'])
print("Beta:", model.model.params.get('smoothing_slope'))
print("Gamma:", model.model.params.get('smoothing_seasonal'))
```

```python
df_set = pd.DataFrame({'DEMANDA':demanda,'FORECAST':set, 'ERROR':demanda-set})
```

```python
kpi(df_set)
```

### 2.4.3 Simulacion 'mas real'

```python
demanda = val_dem_sem[5]
forecast = []
for n in range (105,157):
    values = demanda[:n]
    # Aplica la suavizacion exponencial triple
    model = ExponentialSmoothing(values, seasonal='add', trend='add', seasonal_periods=52).fit()
    set = model.fittedvalues
    forecast_next = model.forecast(steps=1).item()
    forecast.append(forecast_next)
```
